```python
import rasterio
from matplotlib import pyplot

# Відкриття зображення
with rasterio.open("../images/soil_moisture.tif") as dataset:

    # Maсив 
    array = dataset.read(1)
    
    # Bідобразити зображення за допомогою pyplot
    pyplot.imshow(array, cmap='pink')
    pyplot.show()

    # Виведення дійсних і недійсних точок даних
    print(f'Кількість точок які містять правильне значення вологості: {array[array != 65535].size}')
    print(f'Кількість точок з помилками розрахунку: {array[array == 65535].size}')
```
